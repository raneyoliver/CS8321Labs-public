## Plot Histograms and Fit Gamma Distributions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Step 1: Load data from CSV
data = pd.read_csv('iter1_results10000_fin.csv')
values = data['Median_Error'].values  # Adjust the column name if different

# Step 2: Fit the gamma distribution
params = gamma.fit(values)

# Step 3: Plot the histogram and the fitted distribution
fig, ax = plt.subplots()
# Plot histogram of the data
ax.hist(values, bins=30, density=True, alpha=0.5, color='g', label='Data histogram')

# Plot the fitted gamma distribution
x = np.linspace(min(values), max(values), 100)
pdf_fitted = gamma.pdf(x, *params)
ax.plot(x, pdf_fitted, 'r-', label='Fitted gamma distribution')

# Add labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Density')
ax.set_title('Gamma Distribution Fit to Iteration 1')
ax.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# List of datasets
datasets = ['iter1_results10000_fin.csv', 'iter2_results10000_fin.csv', 'iter3_1_bvm1_corr_results10000_fin.csv', 'iter4_1_bvm1_corr_results10000_fin.csv']
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each dataset
labels = ['Iteration 1', 'Iteration 2', 'Iteration 3', 'Iteration 4']

fig, ax = plt.subplots()

# Loop over datasets
for data_file, color, label in zip(datasets, colors, labels):
    # Load data
    data = pd.read_csv(data_file)
    values = data['TotalReward'].values

    # Fit the gamma distribution
    params = gamma.fit(values)
    alpha, loc, scale = params
    mean = gamma.mean(alpha, loc, scale)

    # Plot the fitted gamma distribution
    x = np.linspace(min(values), max(values), 1000)
    pdf_fitted = gamma.pdf(x, alpha, loc, scale)
    ax.plot(x, pdf_fitted, '-', color=color, label=f'{label}: ={alpha:.2f}, ={scale:.2f}, Mean={mean:.2f}')

# Add legend, labels, and title
ax.legend()
ax.set_xscale('log')

#ax.set_xlabel('Median_Error')
ax.set_ylabel('Density')
ax.set_title('Comparison of Gamma Distributions Across Four Datasets')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define dataset files and column types
dataset_files = [
    'iter1_results10000_fin.csv',
    'iter2_results10000_fin.csv',
    'iter3_1_bvm1_corr_results10000_fin.csv',
    'iter4_1_bvm1_corr_results10000_fin.csv'
]
data_columns = {
    'Median_CEP': 'Median CEP',
    'Median_Error': 'Median Error',
    'Num_Steps': 'Number of Steps',
    'Distance_Moved': 'Distance Moved'
}
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each dataset
labels = ['Iteration 1', 'Iteration 2', 'Iteration 3', 'Iteration 4']  # Legend labels

# Setup a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

# Loop through each data type and plot on the corresponding subplot
for ax, (key, label) in zip(axes, data_columns.items()):
    for file, color, leg_label in zip(dataset_files, colors, labels):
        # Load data from each file
        data = pd.read_csv(file)
        values = data[key].values

        # Fit the normal distribution to the specified column
        mu, sigma = norm.fit(values)
        mu, sigma = round(mu), round(sigma)  # Round parameters to the nearest integer

        # Generate x values for plotting
        x = np.logspace(np.log10(max(1, np.min(values))), np.log10(np.max(values)), 1000)
        pdf_fitted = norm.pdf(x, mu, sigma)

        # Plot the fitted normal distribution for each dataset
        ax.plot(x, pdf_fitted, '-', color=color, label=f'{leg_label}: ={mu}, ={sigma}')

        # Set the x-axis to logarithmic scale and dynamically set its limits based on the subplot
        ax.set_xscale('log')
        if key == 'Distance_Moved':
            ax.set_xlim([np.min(values), np.max(values)])
        else:
            ax.set_xlim([1, np.max(values)])

    # Labels and title for each subplot
    #ax.set_title(f'Fit of {label}')
    ax.set_xlabel(label)
    ax.set_ylabel('Density')
    ax.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
