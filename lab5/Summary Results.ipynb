```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

## 1000 Runs

```python
results_iter1 = pd.read_csv('iter1_results1000.csv')
results_iter1.describe().style.format("{:.2f}")
```

```python
results_iter2 = pd.read_csv('iter2_results1000.csv')
results_iter2.describe().style.format("{:.2f}")
```

```python
results_iter3 = pd.read_csv('iter3_1_results1000.csv')
results_iter3.describe().style.format("{:.2f}")
```

```python
results_iter4 = pd.read_csv('iter4_results1000.csv')
results_iter4.describe().style.format("{:.2f}")
```

```python
# Plot Results
plt.scatter(350,350,color='red', label='True Target Location', marker='x',s=200)

mean_num_steps = results_iter1['Num_Steps'].mean()
mean_median_error = results_iter1['Median_Error'].mean()
mean_median_cep = results_iter1['Median_CEP'].mean()
mean_median_x = results_iter1['Median_X'].mean()
mean_median_y = results_iter1['Median_Y'].mean()
mean_mad_x = results_iter1['MAD_X'].mean()
mean_mad_y = results_iter1['MAD_Y'].mean()

# Plot circle using MAD deviations of x & y values
radius = mean_median_cep
median_cep_legend1 = f'Median CEP Iteration 1 \nMean Error: ({np.round(mean_median_error,1)})'
circle_median = Ellipse((mean_median_x, mean_median_y), 2*radius, 2*radius, edgecolor='blue', facecolor='none', linestyle='-', linewidth=2, label=median_cep_legend1)
plt.gca().add_patch(circle_median)

mean_num_steps = results_iter2['Num_Steps'].mean()
mean_median_error = results_iter2['Median_Error'].mean()
mean_median_cep = results_iter2['Median_CEP'].mean()
mean_median_x = results_iter2['Median_X'].mean()
mean_median_y = results_iter2['Median_Y'].mean()
mean_mad_x = results_iter2['MAD_X'].mean()
mean_mad_y = results_iter2['MAD_Y'].mean()

# Plot circle using MAD deviations of x & y values
radius = mean_median_cep
median_cep_legend2 = f'Median CEP Iteration 2 \nMean Error: ({np.round(mean_median_error,1)})'
circle_median = Ellipse((mean_median_x, mean_median_y), 2*radius, 2*radius, edgecolor='red', facecolor='none', linestyle='-', linewidth=2, label=median_cep_legend2)
plt.gca().add_patch(circle_median)

mean_num_steps = results_iter3['Num_Steps'].mean()
mean_median_error = results_iter3['Median_Error'].mean()
mean_median_cep = results_iter3['Median_CEP'].mean()
mean_median_x = results_iter3['Median_X'].mean()
mean_median_y = results_iter3['Median_Y'].mean()
mean_mad_x = results_iter3['MAD_X'].mean()
mean_mad_y = results_iter3['MAD_Y'].mean()

# Plot circle using MAD deviations of x & y values
radius = mean_median_cep
median_cep_legend3 = f'Median CEP Iteration 3 \nMean Error: ({np.round(mean_median_error,1)})'
circle_median = Ellipse((mean_median_x, mean_median_y), 2*radius, 2*radius, edgecolor='black', facecolor='none', linestyle='-', linewidth=2, label=median_cep_legend3)
plt.gca().add_patch(circle_median)


mean_num_steps = results_iter4['Num_Steps'].mean()
mean_median_error = results_iter4['Median_Error'].mean()
mean_median_cep = results_iter4['Median_CEP'].mean()
mean_median_x = results_iter4['Median_X'].mean()
mean_median_y = results_iter4['Median_Y'].mean()
mean_mad_x = results_iter4['MAD_X'].mean()
mean_mad_y = results_iter4['MAD_Y'].mean()

# Plot circle using MAD deviations of x & y values
radius = mean_median_cep
median_cep_legend4 = f'Median CEP Iteration 4 \nMean Error: ({np.round(mean_median_error,1)})'
circle_median = Ellipse((mean_median_x, mean_median_y), 2*radius, 2*radius, edgecolor='green', facecolor='none', linestyle='-', linewidth=2, label=median_cep_legend4)
plt.gca().add_patch(circle_median)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Median CEP by Iteration')
plt.legend(
    loc='lower center', 
    bbox_to_anchor=(0.5, -.35),
    ncol=3, 
)
plt.grid(True)
plt.axis('equal')  # Equal aspect ratio
plt.show()
```
